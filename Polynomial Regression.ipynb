```python
# Let's import important Libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Creation of data fitted for a parabolic function

n_samples = 100
# here we work with a linear model so linpace instead of logspace
# 0 is the beginning and 10 is max with 100 points
X = np.linspace(0, 10, 100) 
# Our equation while we use random.randn to pic random points
y = X ** 3 + np.random.randn(n_samples) * 100 + 100
# delineate space to work in
plt.figure(figsize=(10,8))
plt.title('Distribution of Random points on a parabolic plane')
# we'll work with a scatter plot to show the distribution of the random points
plt.scatter(X, y)
plt.show()
```

```python
# Doing a Linear Regression to compare the fits

lin_reg = LinearRegression()
lin_reg.fit(X.reshape(-1, 1), y)
model_pred = lin_reg.predict(X.reshape(-1,1))
plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, model_pred);
print(r2_score(y, model_pred))
```

Doing the above with Logspace

```python
# Creation of data fitted for a parabolic function

n_samples = 100
# here we work with a linear model so linpace instead of logspace
# 0 is the beginning and 10 is max with 100 points
X = np.logspace(0, 10, 100) 
# Our equation while we use random.randn to pic random points
y = X ** 3 + np.random.randn(n_samples) * 100 + 100
# delineate space to work in
plt.figure(figsize=(10,8))
plt.title('Distribution of Random points on a parabolic plane')
# we'll work with a scatter plot to show the distribution of the random points
plt.scatter(X, y)
plt.show()
```

```python
# Doing a Linear Regression to compare the fits

lin_reg = LinearRegression()
lin_reg.fit(X.reshape(-1, 1), y)
model_pred = lin_reg.predict(X.reshape(-1,1))
plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, model_pred);
print(r2_score(y, model_pred))
```

This is clearly not the right fit for this particular data points

Now let's look at Polynomial Features to process the X

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
p = PolynomialFeatures(degree=2)
f= p.fit_transform(X.reshape(-1, 1))
```

```python
lin_p = LinearRegression()
lin_p.fit(f, y.reshape(-1, 1))
y_pred = lin_p.predict(f)

plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, y_pred);
print(r2_score(y, y_pred))
```

```python
# Let's try to work with a degree=3
p = PolynomialFeatures(degree=3)
f= p.fit_transform(X.reshape(-1, 1))

lin_p = LinearRegression()
lin_p.fit(f, y.reshape(-1, 1))
y_pred = lin_p.predict(f)

plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, y_pred);
print(r2_score(y, y_pred))
```

```python
# Let's try to work with a degree=3
p = PolynomialFeatures(degree=1)
f= p.fit_transform(X.reshape(-1, 1))

lin_p = LinearRegression()
lin_p.fit(f, y.reshape(-1, 1))
y_pred = lin_p.predict(f)

plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, y_pred);
print(r2_score(y, y_pred))
```

Manipulating the degree and perhaps the other polynomial features help with the fit when working with more data.
